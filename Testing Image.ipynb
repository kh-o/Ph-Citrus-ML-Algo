```python
import cv2
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib import colors
from matplotlib import cm
from matplotlib import colors
import numpy as np
```

```python
cal = cv2.imread('cal3.jpg')
plt.imshow(cal)
plt.show()
```

```python
cal = cv2.cvtColor(cal, cv2.COLOR_BGR2RGB)
plt.imshow(cal)
plt.show()
```

```python
figure_size=15
edges = cv2.Canny(cal,100,150)
plt.figure(figsize=(figure_size,figure_size))
plt.subplot(1,2,1),plt.imshow(cal)
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(edges, cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
hsv_cal = cv2.cvtColor(cal, cv2.COLOR_RGB2HSV)
```

```python
green1 = (60, 0, 0)
green2 = (255, 150, 150)
```

```python
mask = cv2.inRange(hsv_cal, green1, green2)
```

```python
result = cv2.bitwise_and(cal, cal, mask=mask)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(mask, cmap="gray")
plt.subplot(1, 2, 2)
plt.imshow(result)
plt.show()
```

```python
cal3 = cv2.imread('cal1.jpg',0)
plt.imshow(cal3)
plt.show()
```

```python
ret, mask1 = cv2.threshold(cal3,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
plt.subplot(1,2,1)
plt.imshow(mask1,cmap="gray")
plt.show()
```

```python
figure_size=15
edges = cv2.Canny(mask1,100,150)
plt.figure(figsize=(figure_size,figure_size))
plt.subplot(1,2,1),plt.imshow(mask1,cmap="gray")
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(edges, cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```
