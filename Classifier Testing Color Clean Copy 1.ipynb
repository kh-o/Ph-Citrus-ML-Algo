## ANN

```python
#Independent variables
input_set = np.array([hmeans, hmeans1, hmeans2, smeans, smeans1])
#Dependent variable
labels = np.array([1,1,1,0,0])
labels = labels.reshape(5,1) #to convert labels to vector, number of inputs to output
```

```python
np.random.seed(42)
weights = np.random.rand(3,1) #size of matrix to 1
bias = np.random.rand(1) 
lr = .001 #learning rate

def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
for train in range(40000):
    inputs = input_set
    XW = np.dot(inputs, weights)+ bias
    z = sigmoid(XW)
    error = z - labels
    print(error.sum())
    dcost = error
    dpred = sigmoid_derivative(z)
    z_del = dcost * dpred
    inputs = input_set.T
    weights = weights - lr*np.dot(inputs, z_del)
    
    for num in z_del:
        bias = bias - lr*num
```

```python
predict = [htmeans, htmeans1, stmeans]

for count in predict:
    single_pt = np.array([count])
    result = sigmoid(np.dot(single_pt, weights) + bias)
    print(result)
```
